```python
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
#                     | S = pow(N, 0)          | S = pow(N, 1)        |    pow(N, 1.5)         |    pow(N, 2)
# (return M[u][v])    || space|threshold|time || space|threshold|time || space|threshold|time || space|threshold|time||
# access = 1000000    |1,41585310,7051ms       | 41585310,6448,7378ms  | 268169828931,80,7362ms|1729338007796100,1,2110ms
# -----------------------------------------------------------------------------------------------------------------------------------------------------------
# Test Output Table   | S = pow(N, 0)          | S = pow(N, 1)        |    pow(N, 1.5)         | pow(N, 1.8)           | pow(N, 2)  
# -----------------------------------------------------------------------------------------------------------------------------------------------------------
# (avoid hash)        || space|threshold|time || space|threshold|time || space|threshold|time || space|threshold|time  || space|threshold|time
# access = 1000000    |1,741053,1584ms        | 741053,860,1580ms    | 637931290,29,1532ms     |36789906626,3,378ms     | 549159548809,1,429ms

# time_lists = [
#     [9,9,7,8,418.035],
#     [64,67,61,67,4046.003],
#     [229,312,283,318,19813.44],
#     [592,618,569,619,38973.425]
# ]
time_lists = [
    [7051,7378,7362,3000,2110,1],
    [1584,1580,1532,429,378,1],
    [211,173,13,13,13,1],
    [1,1,1,1,1,10000],
]
for i in range(len(time_lists)):
    for j in range(len(time_lists[0])):
        time_lists[i][j] = math.log2(time_lists[i][j])
```

```python
figure(figsize=(14,8), dpi=100)

space_keys = ['pow(N, 0)','pow(N, 1) - 860','pow(N, 1.5) - 29','pow(N, 1.8) - 3','pow(N, 2) - 1', 'postgres']# 'postgres'

x_axis = np.arange(len(space_keys))

# plt.bar(x_axis - 0.3, time_lists[0], 0.2, label = '10')
#plt.bar(x_axis - 0.2, time_lists[0], 0.2, label = 'return M[u][v]')
plt.bar(x_axis - 0.2, time_lists[1], 0.4, label = 'egonets dataset')
plt.bar(x_axis + 0.2, time_lists[2], 0.4, label = 'dense  dataset')
plt.bar(x_axis, time_lists[3], 0.4, label = '')
  
plt.xticks(x_axis, space_keys,  fontsize=14)
plt.yticks(fontsize=11)
plt.xlabel("Space",  fontsize=16)
plt.ylabel("log2 (Time Used in ms)", fontsize=16)
plt.title("Test Results",  fontsize=18)
plt.legend(title="",  fontsize=13, loc="upper left")
plt.savefig('results_bar.png')
plt.show()

# plt.legend(pie[0],labels, bbox_to_anchor=(1,0), loc="lower right", bbox_transform=plt.gcf().transFigure)
```
