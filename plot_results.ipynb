```python
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
space_keys = ['pow(N, 0)','pow(N, 0.5)','pow(N, 1)','pow(N, 1.5)','pow(N, 2)']# 'postgres'
```

```python
# N = 5000
intersection_time_lists_egonets = [
    [72671,72534,72645,12,12], # egonets 10000
    [795,795,796,396,182], #  synthetic 2000
#     [2485,2484,2481,2478,2464,2449, 1], # synthetic 1000
#     [7000,6997,6985,6982,6949,6901, 1], # synthetic 2000
#     [1,   1,   1,   1,   1,    1, 5573416.460],# postgres 1000
#     [1,   1,   1,   1,   1,    1, 12087387.932]# postgres 2000
]

for i in range(len(intersection_time_lists_egonets)):
    for j in range(len(intersection_time_lists_egonets[0])):
        intersection_time_lists_egonets[i][j] = math.log2(intersection_time_lists_egonets[i][j])

        
figure(figsize=(6,4), dpi=100)

x_axis = np.arange(len(space_keys))

plt.bar(x_axis - 0.2, intersection_time_lists_egonets[0], 0.4, label = 'egonets 10000 access')
plt.bar(x_axis + 0.2, intersection_time_lists_egonets[1], 0.4, label = 'synthetic 2000 access')
# plt.bar(x_axis + 0.05, intersection_time_lists[2], 0.1, label = 'synthetic 1000 access')
# plt.bar(x_axis + 0.15, intersection_time_lists[3], 0.1, label = 'synthetic 2000 access')

# plt.bar(x_axis - 0.2, intersection_time_lists_egonets[2], 0.4, label = 'postgres 1000 access')
# plt.bar(x_axis + 0.2, intersection_time_lists_egonets[3], 0.4, label = 'postgres 2000 access')

  
plt.xticks(x_axis, space_keys,  fontsize=14)
plt.yticks(fontsize=11)
plt.xlabel("Space",  fontsize=16)
plt.ylabel("log2 (Time Used in ms)", fontsize=16)
plt.title("2-Set-Intersection",  fontsize=18)
plt.legend(title="",  fontsize=13, loc="upper right")
for pos in ['right', 'top']: plt.gca().spines[pos].set_visible(False)
plt.savefig('2-Set-Intersection.png')
plt.show()

# plt.legend(pie[0],labels, bbox_to_anchor=(1,0), loc="lower right", bbox_transform=plt.gcf().transFigure)
```
