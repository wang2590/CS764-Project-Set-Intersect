```python
import math
```

```python
# access = 10         | 27421,165,9ms         | 588177,35,9ms         | 12616336,7,7ms        | 751911241,1,8ms      | 418.035ms
# access = 100        | 27421,165,64ms        | 588177,35,67ms        | 12616336,7,61ms       | 751911241,1,67ms     | 4046.003ms
# access = 500        | 27421,165,299ms       | 588177,35,312ms       | 12616336,7,283ms      | 751911241,1,318ms    | 19813.44ms
# access = 1000       | 27421,165,592ms       | 588177,35,618ms       | 12616336,7,569ms      | 751911241,1,619ms    | 38973.425ms

time_lists = [
    [9,9,7,8,418.035],
    [64,67,61,67,4046.003],
    [229,312,283,318,19813.44],
    [592,618,569,619,38973.425]
]
for i in range(4):
    for j in range(5):
        time_lists[i][j] = math.log2(time_lists[i][j])
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

figure(figsize=(12,8), dpi=300)

space_keys = ['pow(N, 1)','pow(N, 1.3)','pow(N, 1.6)','pow(N, 2)','postgres']

x_axis = np.arange(len(space_keys))

plt.bar(x_axis - 0.3, time_lists[0], 0.2, label = '10')
plt.bar(x_axis - 0.1, time_lists[1], 0.2, label = '100')
plt.bar(x_axis + 0.1, time_lists[2], 0.2, label = '500')
plt.bar(x_axis + 0.3, time_lists[3], 0.2, label = '1000')
  
plt.xticks(x_axis, space_keys)
plt.xlabel("Space")
plt.ylabel("log2 (Time Used in ms)")
plt.title("Test Results")
plt.legend(title="Num Access")
plt.savefig('results_bar.png')
plt.show()
```
